```python
## 2. When does the code in a function run: when it's specified or when it's called?
'''
Function Definition: When you define a function, you provide a name for the function, specify its parameters (if any),
                     and define the code that the function should execute when called. However, this code is not executed 
                     at the moment of definition.

Function Call: When you want to execute the code within the function, you call (invoke) the function by using its name 
               followed by parentheses containing any required arguments. This triggers the execution of the function's code.

Code Execution: Upon calling the function, the program jumps to the location of the function definition and begins 
                executing the statements within the function's block of code. The parameters receive the values 
                provided during the function call, and any operations within the function are performed.

'''
```

```python
## 6.What happens to variables in a local scope when the function call returns?
'''
     When a function call returns, the local variables that were defined within the function's local scope are deallocated and 
     their values are discarded. The local scope is destroyed, and any variables that were only accessible within that scope are
     no longer accessible or valid.
'''
```

```python
## 13.What can you do to save a programme from crashing if it encounters an error?
'''
   To prevent a program from crashing when it encounters an error, you can implement error handling mechanisms in your code.
   
   1. Try...Except Blocks (Exception Handling): Use try...except blocks to catch and handle exceptions. This allows you to 
      specify code that should run normally (in the try block) and code that should execute when an exception occurs 
      (in the except block). By catching exceptions, you can prevent them from causing the program to crash.
      
      Example:
               try:
                    # Code that may raise an exception
                    result = 20 / 0  # This will raise a ZeroDivisionError
               except ZeroDivisionError as E:
                    # Handle the exception
                    print(f"An error occurred: {E}")
                    result = None 
                    
    2. Logging: Implement a logging mechanism to record errors and other relevant information. 
    
       Example:
                import logging

                try:
                    # Code that may raise an exception
                except Exception as e:
                    # Log the error
                    logging.error(f"An error occurred: {e}")
'''
```

```python
## 14.What is the purpose of the try clause? What is the purpose of the except clause?
'''
    1.Try Clause:
                The try clause is used to enclose a block of code that may potentially raise exceptions. Its primary purpose 
                is to define a "try" block in which you place the code that you want to monitor for exceptions.
                
    2.Except Clause:
                The except clause is used to specify what should happen when a particular type of exception occurs within 
                the preceding try block.
                
    Example:
               try:
                   result = 10 / 0  # This will raise a ZeroDivisionError
               except ZeroDivisionError as e:
               # Handle the ZeroDivisionError exception
                   print(f"An error occurred: {e}")
                   
    >If an exception matching the type specified in an except clause occurs in the try block, 
     the code in that except block is executed.
    
    >If no exception occurs, the except block is skipped, and the program continues after the try...except block.
                
'''
```
