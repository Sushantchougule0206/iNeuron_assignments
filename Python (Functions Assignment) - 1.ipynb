```python
## 3.What is the purpose of the return statement in a function? Can a function have multiple return statements? Explain with an example.

'''
    The return statement in a function serves the purpose of specifying the value(s) that the function should produce as output 
    or result when it is called.
    
    The return statement can also terminate the execution of the function, and any code after the return statement will not be 
    executed.

    A function can indeed have multiple return statements. However, once a return statement is executed, the function's 
    execution is terminated, and the specified value is returned.
    
    Ex: def mul_num(num1,num2):
            mul = num1 * num2
            return mul
        result = mul_num(8,3)
        print("Multiplication of two numbers is",result)
'''
```

```python
## 4.What are lambda functions in Python? How are they different from regular functions? Provide an example where a lambda function can be useful.

'''
   Lambda functions, also known as anonymous functions or lambda expressions.They are defined using the lambda keyword and are 
   typically used for simple, one-line operations.
   
   Differences between Lambda Functions and Regular Functions:
   
    1. Lambda functions have no name or identifier, whereas regular functions defined with def have a name.

    2. Lambda functions use a compact syntax with the lambda keyword, while regular functions use the def keyword.
    
    3. Lambda functions are typically used for simple, single-line operations, while regular functions can contain multiple 
       lines of code and are suitable for more complex logic.
'''

list_square = lambda x: [i ** 2 for i in x]
print(list_square([1,2,3,4,5,6]))
```

```python
## 5.How does the concept of "scope" apply to functions in Python? Explain the difference between local scope and global scope.
 
''' The concept of "scope" refers to the region of the program where a particular variable is visible and can be accessed. 
    Python uses a hierarchical structure to define scopes,
    which can be broadly categorized into two main types: local scope and global scope.

    Local Scope: 
           1. Local scope, also known as function scope, is the most limited scope in Python.
           
           2. Variables defined within a function are considered to have local scope. These variables are known as "local variables."
           
           Ex: def my_function():
                   x = 10  
                   print(x)

               my_function()
               
    Global Scope: 
           1. Global scope is the broader scope that encompasses the entire Python program or module.
           
           2. Variables defined outside of any function or at the top level of a module have global scope. These variables are 
              known as "global variables."
              
           Ex: global_variable = 20  # This is a global variable with global scope

               def my_function():
                   global global_variable
                   print(global_variable)  # Accessing the global variable

               my_function()
'''
```

```python
## 7.What is the difference between the "pass by value" and "pass by reference" concepts when it comes to function arguments in Python?

''' 1. Pass by Value:
            In "pass by value," a copy of the actual data (the value) is passed to the function. Changes made to the function's 
            parameter inside the function do not affect the original variable in the caller's scope.

            Ex: def modify_value(x):
                    x = x + 1

                value = 5
                modify_value(value)
                print(value) 
                
                
    2. Pass by Reference (Traditional):
             In "pass by reference," a reference to the original data (the memory location) is passed to the function. Any 
             changes made to the parameter inside the function directly affect the original variable in the caller's scope.
             
             Ex: def modify_list(my_list):
                     my_list.append(4) 

                 my_list = [1, 2, 3]
                 modify_list(my_list)
                 print(my_list) 
'''
```

```python
## 8.Create a function that can intake integer or decimal value and do following operations:
'''a. Logarithmic function (log x)
   b. Exponential function (exp(x))
   c. Power function with base 2 (2^x)
   d. Square root'''

import math

def math_operations(x):
    result = {}
    
    result['logarithm'] = math.log(x)
    
    result['exponential'] = math.exp(x)
    
    result['power_2'] = 2 ** x
    
    result['square_root'] = math.sqrt(x)
    
    return result

value = 4.0  
results = math_operations(value)

print(f"Input Value: {value}")
print(f"Logarithmic Function (log x): {results['logarithm']}")
print(f"Exponential Function (exp(x)): {results['exponential']}")
print(f"Power Function with Base 2 (2^x): {results['power_2']}")
print(f"Square Root: {results['square_root']}")
```

```python
## 9.Create a function that takes a full name as an argument and returns first name and last name.
def extract_names(full_name):
    # Split the full name into a list of words
    name_parts = full_name.split()

    # Check if there are at least two words (first name and last name)
    if len(name_parts) >= 2:
        # The first name is the first word in the list
        first_name = name_parts[0]

        # The last name is the last word in the list
        last_name = name_parts[-1]

        return (first_name, last_name)
    else:
        
        return None

full_name = str(input("Enter Your Name"))
first, last = extract_names(full_name)
if first and last:
    print("First Name:", first)
    print("Last Name:", last)
else:
    print("Invalid full name format.")
```
